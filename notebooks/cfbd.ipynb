```python
import os
import requests

API = 'https://api.collegefootballdata.com'

def get(resource):
    return requests.get(
        f'{API}{resource}',
        headers={
            'Authorization': f'Bearer {os.getenv("CFBD_API_KEY")}'
        },
    )
```

```python
res = get('/games?year=2025')
games = res.json()
games
```

```python
box_scores_by_week = []

for week in range(1, 14):
    res = get(f'/games/teams?year=2025&week={week}')
    box_scores = res.json()
    box_scores_by_week.append(box_scores)
    
box_scores_by_week
```

```python
res = get(f'/lines?year=2025')
lines = res.json()
lines
```

```python
import pandas as pd

games_df = pd.DataFrame(games)
games_df
```

```python
lines_df = pd.DataFrame(lines)
lines_df
```

```python
games_df.to_csv('cfbd_2025_games.csv', index=False)
lines_df.to_csv('cfbd_2025_lines.csv', index=False)
```
